## 5. Carga — Hugging Face `Shomi28/cyber-threat-intelligence`

El *dataset* en el Hub declara un builder único que **no es reproducible** con `load_dataset("…")` solo (mezcla de Parquets con esquemas distintos). Aquí se cargan los **cuatro archivos Parquet publicados** (`cve_data`, `mitre_attack`, instrucciones train/test) mediante rutas `hf://datasets/...`.


```python
from datasets import load_dataset

HF_ID = "Shomi28/cyber-threat-intelligence"
HF_BASE = f"hf://datasets/{HF_ID}"

# `load_dataset(HF_ID)` en el Hub actual suele FALLAR: un solo builder mezcla Parquets con
# esquemas distintos (CVE vs instruction tuning). Carga estable = rutas Parquet explícitas.

parquet_map = {
    "cve_data": f"{HF_BASE}/cve_data/train-00000-of-00001.parquet",
    "mitre_attack": f"{HF_BASE}/mitre_attack/train-00000-of-00001.parquet",
    "instructions_train": f"{HF_BASE}/instructions/train-00000-of-00001.parquet",
    "instructions_test": f"{HF_BASE}/instructions/test-00000-of-00001.parquet",
}

print("Cargando Hugging Face (Parquet por bloque):", HF_ID)
hf_datasets = {}
for name, url in parquet_map.items():
    ds = load_dataset("parquet", data_files=url)
    hf_datasets[name] = ds
    print(f"  [{name}] splits: {list(ds.keys())}")

# Resumen: filas por split y ejemplo mínimo
for cfg_name, dset in hf_datasets.items():
    print(f"\n=== Bloque: {cfg_name} ===")
    for split_name, split in dset.items():
        n = len(split)
        print(f"  {split_name}: {n} filas")
        print(f"  features: {split.features}")
        if n > 0:
            print("  ejemplo[0] keys:", list(split[0].keys()))
            sample = split[0]
            preview = {
                k: (str(v)[:120] + "…" if isinstance(v, str) and len(str(v)) > 120 else v)
                for k, v in list(sample.items())[:6]
            }
            print("  ejemplo[0] (parcial):", preview)
```

## 6. Carga — Kaggle `prayanshusinghgaur/cybersecurity-threat-intelligence`

Dataset **público**:


```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
from pathlib import Path

import pandas as pd

KAGGLE_HANDLE = "prayanshusinghgaur/cybersecurity-threat-intelligence"

_load = getattr(kagglehub, "load_dataset", None)
_df = None
if _load is not None:
    try:
        _df = _load(KaggleDatasetAdapter.PANDAS, KAGGLE_HANDLE)
    except TypeError:
        pass  # muchas versiones exigen el argumento path

if _df is None:
    _dataset_load = getattr(kagglehub, "dataset_load", None)
    if _dataset_load is None:
        raise ImportError("Instala o actualiza kagglehub[pandas-datasets]")
    root = Path(kagglehub.dataset_download(KAGGLE_HANDLE))
    csvs = sorted(root.rglob("*.csv"))
    if not csvs:
        raise FileNotFoundError(f"No hay .csv en {root}")
    rel = csvs[0].relative_to(root).as_posix()
    print(f"kagglehub: usando archivo {rel!r}")
    _df = _dataset_load(KaggleDatasetAdapter.PANDAS, KAGGLE_HANDLE, rel)

df_kaggle = _df
if isinstance(df_kaggle, dict):
    df_kaggle = next(iter(df_kaggle.values()))

print("Primeras filas:")
display(df_kaggle.head())
print("shape:", df_kaggle.shape)
print(df_kaggle.dtypes)
```

```python
# Diccionario inferido (tabla) — dataset Kaggle cargado
null_pct = (df_kaggle.isna().mean() * 100).round(2)
examples = []
for col in df_kaggle.columns:
    val = df_kaggle[col].dropna().iloc[0] if df_kaggle[col].notna().any() else None
    s = str(val)
    examples.append(s[:80] + ("…" if len(s) > 80 else ""))

dict_kaggle_df = pd.DataFrame(
    {
        "columna": df_kaggle.columns,
        "dtype": df_kaggle.dtypes.astype(str).values,
        "%_nulos": null_pct.values,
        "ejemplo": examples,
    }
)
display(dict_kaggle_df)
```
